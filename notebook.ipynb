```python
import pandas as pd
```

```python
df_trips = pd.read_parquet("data/green_tripdata_2025-11.parquet", engine="pyarrow")
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
df_trips.to_sql(
    name="trip_data",
    con=engine,
    if_exists="replace"
)
```

```python
dtypes ={
    "LocationID": "Int64",
    "Borough": "category",
    "Zone": "string",
    "service_zone": "category"
}
```

```python
df_zones = pd.read_csv("data/taxi_zone_lookup.csv", dtype=dtypes)
df_zones = df_zones.set_index("LocationID") 
```

```python
df_zones
```

```python
df_zones.to_sql(
    name="zones",
    con=engine,
    index=True,
    index_label="LocationID",
    if_exists="replace"
)
```
